```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("properties.csv")
df.head()
df.info()
df.describe()
```

### Sale Price 

```python
plt.figure(figsize=(8,4))
sns.histplot(df['SalePrice'], bins=30, kde=True)
plt.title('Distribution of Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

### Size_sqft


```python
plt.figure(figsize=(8,4))
sns.histplot(df['Size_sqft'], bins=30, kde=True, color='green')
plt.title('Distribution of Property Size (sqft)')
plt.xlabel('Size (sqft)')
plt.ylabel('Frequency')
plt.show()
```

### Location

```python
plt.figure(figsize=(4,4))
sns.boxplot(x='Location', y='SalePrice', data=df)
plt.title('Sale Price by Location')
plt.xticks(rotation=45)
plt.show()
```

### Heat Map

```python

numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Price vs Size

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='Size_sqft', y='SalePrice', data=df, hue='Location')
plt.title('Sale Price vs Property Size')
plt.xlabel('Size (sqft)')
plt.ylabel('Sale Price')
plt.show()
```
